### Split dataset

```python
X = df_test.drop('symboling', axis=1)
X.shape
```

```python
y = df_test['symboling']
y.shape
```

```python
# karena total data hanya 159
# maka kita bagi saja menjadi 80:20

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2, shuffle=True) 
```

```python
X_train.shape
```

```python
X_test.shape
```

### Train data

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train) 
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# R-Square menunjukkan seberapa baik variabel independen (input)
# memprediksi variabel dependen (output).
r2 = r2_score(y_test, y_pred)
print(f"R-square: {r2:.4f}")

# MSE menghitung selisih antara nilai prediksi model dan nilai sebenarnya dari data
# metrik evaluasi yang digunakan untuk mengukur akurasi sebuah model regresi dalam memprediksi output.
mse = mean_squared_error(y_test, y_pred)
print(f"Mean squared error: {mse:.4f}")

# RMSE menghitung besarnya rata-rata kesalahan antara data hasil prediksi dengan data aktual.
rmse = mse ** 0.5
print(f"Root mean squared error: {rmse:.4f}")
```
